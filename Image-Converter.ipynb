```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image, ImageFilter
```

```python
def image_to_emnist(img):
    img = img.resize((28, 28), Image.ANTIALIAS).filter(ImageFilter.SHARPEN)
    
    img_canvas = Image.new('L', (28, 28), (255))
    img_canvas.paste(img)
    
    img_data = list(img_canvas.getdata())

    #Normalize pixels to 0 and 1. 0 is pure white, 1 is pure black.
    img_normalized = [(255 - x) * 1.0 / 255.0 for x in img_data]
    
    return img_normalized
```

```python
symbol_image = Image.open('./sample-images/B-centered-squared.jpg')

plt.imshow(symbol_image, cmap='Greys_r')
print('Image size', symbol_image.size)
```

```python
converted_image = np.array(image_to_emnist(symbol_image))
converted_image.shape
```

```python
converted_image = np.array(converted_image).reshape(28,28)
converted_image.shape
```
